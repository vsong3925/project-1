### 1) API: Requests URL & Create CSV File

#### API: American Community Survey 1-Year Data (2005-2019) - State Level

```python
# Create csv file for individual year from 2009 to 2019

# Means of Transportation
years = ['2009','2010','2011','2012','2013','2014','2015','2016','2017','2018','2019']
variable = 'B08101_001E'
title = 'transportation'

for year in years:
    url = f'https://api.census.gov/data/{year}/acs/acs1?get=NAME,{variable}&for=state:*&key={api_key}'
    data = requests.get(url).json()
    col_name = ['state', year, 'state code']
    df = pd.DataFrame(data[1:], columns=col_name)
    clean_df = df.drop(columns=['state code'])
    clean_df.to_csv(f'{year}_{title}.csv')
```

```python
# Worked From Home
years = ['2009','2010','2011','2012','2013','2014','2015','2016','2017','2018','2019']
variable = 'B08101_049E'
title = 'worked_fm_home'

for year in years:
    url = f'https://api.census.gov/data/{year}/acs/acs1?get=NAME,{variable}&for=state:*&key={api_key}'
    data = requests.get(url).json()
    col_name = ['state', year, 'state code']
    df = pd.DataFrame(data[1:], columns=col_name)
    clean_df = df.drop(columns=['state code'])
    clean_df.to_csv(f'{year}_{title}.csv')
```

```python
# Public Transporation + Taxi/Bikes
year = ['2009','2010','2011','2012','2013','2014','2015','2016','2017','2018','2019']
# Population
va_public_transportation = dmv_public_transportation['VA Public Transportation']
md_public_transportation = dmv_public_transportation['MD Public Transportation']

# Expenditure
va_tservice = pd.DataFrame(va_data.iloc[16])
md_tservice = pd.DataFrame(md_data.iloc[16])

# Plot
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2)
ax1.plot(year, va_public_transportation, color = 'coral', linestyle = '--', label='VA')
ax1.plot(year, md_public_transportation, color = 'steelblue', linestyle = '--', label='MD')

ax2.plot(year, va_tservice, color = 'coral', label='VA')
ax2.plot(year, md_tservice, color = 'steelblue', label='MD')

ax1.legend()
ax1.set_title('Population: TS')
ax1.set_ylabel('Population')

ax2.legend()
ax2.set_title('Expenditure: TS')
ax2.set_ylabel('Million Dollars')

plt.setp((ax1, ax2), xticks=[0, 5, 10], xticklabels=['2009', '2015', '2019'])

plt.tight_layout()
plt.show()

# fig.savefig('Expenditure/Population_Public Transportation.png')
```

### Average Compare

```python
# average expense on public transporation in VA
# average = expense/population
# pie chart: which state is higher
# VA
va_public_transportation = dmv_public_transportation['VA Public Transportation']
va_tservice = pd.DataFrame(va_data.iloc[16])
va = pd.concat([va_public_transportation,va_tservice], axis=1)

expenditure = va.iloc[:,1]
population = va.iloc[:,0]
va['average'] = expenditure / population
va_average = va.round(4)

#MD
md_public_transportation = dmv_public_transportation['MD Public Transportation']
md_tservice = pd.DataFrame(md_data.iloc[16])
md = pd.concat([md_public_transportation,md_tservice], axis=1)

expenditure = md.iloc[:,1]
population = md.iloc[:,0]
md['average'] = expenditure / population
md_average = md.round(4)

# Plot
fig, ax1 = plt.subplots()
ax1.plot(year, va_average, color = 'coral', linestyle = '--', label='VA')
ax1.plot(year, md_average, color = 'steelblue', linestyle = '--', label='MD')


ax1.legend()
ax1.set_title('Population: TS')
ax1.set_ylabel('Population')

#ax2.legend()
#ax2.set_title('Expenditure: TS')
#ax2.set_ylabel('Million Dollars')

plt.setp((ax1), xticks=[0, 5, 10], xticklabels=['2009', '2015', '2019'])

plt.tight_layout()
plt.show()
```

```python
# DMV Walked to Work Population Over 10 Years
year = ['2009','2010','2011','2012','2013','2014','2015','2016','2017','2018','2019']

va_dmv_walked = dmv_walked['VA Walked']
md_dmv_walked = dmv_walked['MD Walked']
dc_dmv_walked = dmv_walked['DC Walked']

# Plot
fig, (ax1, ax2, ax3) = plt.subplots(nrows=3, ncols=1)  # sharex=True
ax1.plot(year, va_dmv_walked, color = 'coral', linestyle = '--', label='VA')
ax2.plot(year, md_dmv_walked, color = 'grey', linestyle = '--', label='MD')
ax3.plot(year, dc_dmv_walked, color = 'steelblue', linestyle = '--', label='DC')

ax1.legend()
ax1.set_title('Population for Walked to Work')
ax1.set_ylabel('Population')

ax2.legend()
ax2.set_ylabel('Population')

ax3.legend()
ax3.set_xlabel('Years')
ax3.set_ylabel('Population')

plt.tight_layout()
plt.show()

# fig.savefig('DMV Walked to Work Population.png')
```
